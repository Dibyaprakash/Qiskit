```python
from qiskit import *
from qiskit.visualization import plot_bloch_multivector
from qiskit.visualization import plot_state_qsphere
%matplotlib inline
from math import pi
#T gate
qc = QuantumCircuit(1,1,name = "qc")
qc.x(0)
qc.t(0)
qc.draw(output = 'mpl')
```

```python
#Execute Quantum Circuit
backend = Aer.get_backend('statevector_simulator')
result = execute(qc,backend).result()
stateVectorResult = result.get_statevector(qc, decimals=3)
print("\n Qunatum State is :",stateVectorResult)
```

```python
plot_bloch_multivector(stateVectorResult)
```

```python
#Unitary Operator
backend = Aer.get_backend('unitary_simulator')
unitary = execute(qc,backend)
unitary.result().get_unitary(qc,decimals=3)
```

```python
plot_state_qsphere(stateVectorResult)
```
